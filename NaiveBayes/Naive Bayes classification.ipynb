```python
# Required Python Machine learning Packages
import pandas as pd
import numpy as np
# For preprocessing the data
from sklearn.preprocessing import Imputer
from sklearn.model_selection import train_test_split
# function for transforming documents into counts
from sklearn.feature_extraction.text import CountVectorizer
# function for encoding categories
from sklearn.preprocessing import LabelEncoder
from sklearn import preprocessing
from sklearn.naive_bayes import MultinomialNB
# To split the dataset into train and test datasets
from sklearn.cross_validation import train_test_split
# To model the Gaussian Navie Bayes classifier
from sklearn.naive_bayes import GaussianNB
# To calculate the accuracy score of the model
from sklearn.metrics import accuracy_score
```

```python
csgo_df = pd.read_csv('counter-strike/playerStats.csv', engine='python')
```

```python
#Handling missing data
csgo_df.isnull().sum()
```

```python
csgo_df_rev = csgo_df
```

```python
#summary statistics of all the attributes.
csgo_df_rev.describe(include= 'all')
```

```python
for value in ['Map','Player','Kills', 'Deaths', 'KAST', 'Rating','MatchID']:
    csgo_df_rev[value].replace([''], [csgo_df_rev.describe(include='all')[value][2]],
                                inplace=True)
```

```python
csgo_df_rev
```

```python
#Encoding the data into binary format
le = preprocessing.LabelEncoder()
Map_cat = le.fit_transform(csgo_df.Map)
Player_cat   = le.fit_transform(csgo_df.Player)
Kills_cat = le.fit_transform(csgo_df.Kills)
Deaths_cat = le.fit_transform(csgo_df.Deaths)
#KAST_cat = le.fit_transform(csgo_df.KAST)
Rating_cat = le.fit_transform(csgo_df.Rating)
MatchID_cat = le.fit_transform(csgo_df.MatchID)
```

```python
#initialize the encoded categorical columns
csgo_df_rev['Map_cat'] = Map_cat
csgo_df_rev['Player_cat'] = Player_cat
csgo_df_rev['Kills_cat'] = Kills_cat
csgo_df_rev['Deaths_cat'] = Deaths_cat
csgo_df_rev['KAST_cat'] = KAST_cat
csgo_df_rev['Rating_cat'] = Rating_cat
csgo_df_rev['MatchID_cat'] = MatchID_cat
```

```python
#drop the old categorical columns from dataframe
dummy_fields = ['Map','Player','Kills', 'Deaths', 'ADR', 'Rating','MatchID']
csgo_df_rev = csgo_df_rev.drop(dummy_fields, axis = 1)
```

```python
csgo_df_rev.head()
```

```python
#Reindexing the column

csgo_df_rev = csgo_df_rev.reindex_axis(['Map_cat','Player_cat','Kills_cat', 'Deaths_cat', 'KAST_cat', 'Rating_cat','MatchID_cat'],axis=1)

csgo_df_rev.head(1)
```

```python
#Standardization of Data
num_features = ['Map_cat','Player_cat','Kills_cat', 'Deaths_cat', 'KAST_cat', 'Rating_cat','MatchID_cat']

scaled_features = {}
for each in num_features:
    mean, std = csgo_df_rev[each].mean(), csgo_df_rev[each].std()
    scaled_features[each] = [mean, std]
    csgo_df_rev.loc[:, each] = (csgo_df_rev[each] - mean)/std
```

```python
#Data Slicing

X = np.asarray(csgo_df[['Player', 'Deaths', 'Rating']])
y = np.asarray(csgo_df['Kills'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.40, random_state = 10)
```

# Gaussian Navie Bayes Implementation

```python
gnb = GaussianNB()


# Train our classifier
model = gnb.fit(X_train, y_train)
```

```python
preds = gnb.predict(X_test)
print(preds)
```

```python
from sklearn.metrics import accuracy_score

# Evaluate accuracy
print(accuracy_score(y_test, preds))
```
